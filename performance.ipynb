```python
import psutil
import os
import matplotlib.pyplot as plt
from psutil import cpu_percent
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib notebook

frame_lenght = 2000
y = []
process = psutil.Process();
fig1 = plt.figure(figsize = (6,2))

def animate_cpu(i):
    y.append(process.cpu_percent())
    if len(y) <= frame_lenght:
        plt.cla()
        plt.plot(y, 'r', label = 'Real-Time CPU %')
    else:
        plt.cla()
        plt.plot(y[-frame_lenght:], 'r', label = 'Real-Time CPU %')
        
    
    #plt.ylim(0, 100)
    plt.xlabel('Time(s)')
    plt.ylabel('CPU(%)')
    #plt.legend(loc = 'upper right')
    plt.tight_layout()
    
ani1 = FuncAnimation(fig1, animate_cpu, interval = 1000)
```

```python
%matplotlib notebook

frame_lenght = 2000
y = []
process = psutil.Process();
fig2 = plt.figure(figsize = (6,2))

def animate_memory(i):
    y.append(process.memory_percent())
    if len(y) <= frame_lenght:
        plt.cla()
        plt.plot(y, 'b', label = 'Real-Time Memory %')
    else:
        plt.cla()
        plt.plot(y[-frame_lenght:], 'r', label = 'Real-Time CPU %')
        
    plt.xlabel('Time(s)')
    plt.ylabel('Memory(%)')
    #plt.legend(loc = 'upper right')
    plt.tight_layout()
    
ani2 = FuncAnimation(fig2, animate_memory, interval = 1000)
```
